```python
import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelWithLMHead
from datasets import load_dataset
from collections import defaultdict
import pandas as pd
from tqdm import tqdm
import pandas as pd
import nltk
from bert_score import score
from torchmetrics.text.rouge import ROUGEScore
from comet import download_model, load_from_checkpoint
from SEScore2 import SEScore2
from train.regression import *
import MeCab
import pyidaungsu as pds
from sklearn.model_selection import train_test_split
import random
from indicnlp import common
from indicnlp import loader
from indicnlp.tokenize import indic_tokenize
from utoken import utokenize




nltk.download('punkt')
nltk.download('wordnet')
```

**Multilingual Summarization Evaluation Data**

```python
WHITESPACE_HANDLER = lambda k: re.sub('\s+', ' ', re.sub('\n+', ' ', k.strip()))

# multilingual baseline model
multi_model_name = "csebuetnlp/mT5_multilingual_XLSum"
multi_tokenizer = AutoTokenizer.from_pretrained(multi_model_name)
multi_model = AutoModelForSeq2SeqLM.from_pretrained(multi_model_name).to('cuda:0')

# generation function
def multilingual_summarize(article_text):
  input_ids = multi_tokenizer(
      [WHITESPACE_HANDLER(article_text)],
      return_tensors="pt",
      padding="max_length",
      truncation=True,
      max_length=512
  )["input_ids"].to("cuda:0")

  output_ids = multi_model.generate(
      input_ids=input_ids,
      max_length=84,
      no_repeat_ngram_size=2,
      num_beams=4
  )[0]

  summary = multi_tokenizer.decode(
      output_ids,
      skip_special_tokens=True,
      clean_up_tokenization_spaces=False
  )

  return summary

languages = ['swahili', 'tamil', 'welsh', 'french', 'punjabi', 'burmese']

results = defaultdict(lambda: defaultdict(list))


for language in languages:
  dataset = load_dataset('csebuetnlp/xlsum', language)
  for row in tqdm(dataset['test']):
    results[language]['generated'].append(multilingual_summarize(row['text']))
    results[language]['ground_truth'].append(row['summary'])

  pd.DataFrame(results[language]).to_csv('/content/drive/MyDrive/MNLP Project/Data/' + language + '.csv', index = False)
```

**Monolingual Summarization as Benchmark to Compare**

```python
mono_model_name = "mrm8488/t5-base-finetuned-summarize-news"
mono_tokenizer = AutoTokenizer.from_pretrained(mono_model_name)
mono_model = AutoModelWithLMHead.from_pretrained(mono_model_name).to('cuda:0')

def monolingual_summarize(text):
  input_ids = mono_tokenizer.encode(text, return_tensors="pt", add_special_tokens=True).to('cuda:0')
  generated_ids = mono_model.generate(
      input_ids=input_ids,
      max_length=84,
      no_repeat_ngram_size=2,
      num_beams=4
  )
  preds = [mono_tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in generated_ids]
  return preds[0]


  results = defaultdict(lambda: defaultdict(list))


for language in languages:
  dataset = load_dataset('csebuetnlp/xlsum', language)
  for row in tqdm(dataset['test']):
    results[language]['generated'].append(monolingual_summarize(row['text']))
    results[language]['ground_truth'].append(row['summary'])

for language in results:
  pd.DataFrame(results[language]).to_csv('/content/drive/MyDrive/MNLP Project/Data/' + language + '-monolingual.csv', index = False)
```

**Calculating Metrics**

```python
df_final = None
langauges = ['tamil', 'welsh', 'burmese', 'french', 'punjabi', 'swahili', 'japanese']
language_codes = {'tamil' : 'ta', 'welsh' : 'cy', 'burmese' : 'my', 'french' : 'fr', 'punjabi' : 'pa', 'swahili' : 'sw', 'japanese' : 'ja'}
comet_path = download_model("Unbabel/wmt22-comet-da")
comet = load_from_checkpoint(comet_path)
scorer = SEScore2('ja')

for language in langauges:
  df = pd.read_csv('/content/drive/MyDrive/MNLP Project/Data/' + language + '.csv')
  df['language'] = language
  df['language_code'] = language_codes[language]
  gen_list = df.generated.tolist()
  gt_list = df.ground_truth.tolist()
  bscore = score(gen_list, gt_list, lang=language_codes[language], verbose = True)
  df['bert_score_precision'] = pd.Series(bscore[0].tolist())
  df['bert_score_recall'] = pd.Series(bscore[1].tolist())
  df['bert_score_f1'] = pd.Series(bscore[2].tolist())
  comet_input = []
  sescore2_input = ([], [], 1)
  for temp in zip(gt_list, gen_list):
    comet_input.append({'src': temp[0], 'mt' : temp[1], 'ref' : temp[0]})
    sescore2_input[0].append(temp[0])
    sescore2_input[1].append(temp[1])
  df['comet_score'] = comet.predict(comet_input)['scores']
  if language == 'japanese':
    df['sescore2_score'] = scorer.score(sescore2_input[0], sescore2_input[1], sescore2_input[2])
  else:
    df['sescore2_score'] = -1

  if df_final is None:
     df_final = df
  else:
    df_final = pd.concat([df_final, df])

rouge = ROUGEScore()
df_final['rouge_scores'] = df_final.apply(lambda x: rouge(x.generated, x.ground_truth), axis = 1)
df_final['meteor_score'] = df_final.apply(lambda x: nltk.translate.meteor_score.meteor_score([nltk.tokenize.word_tokenize(x.generated)], nltk.tokenize.word_tokenize(x.ground_truth)), axis = 1)
df = df_final.reset_index()

rouges = ['rouge1_fmeasure',
 'rouge1_precision',
 'rouge1_recall',
 'rouge2_fmeasure',
 'rouge2_precision',
 'rouge2_recall',
 'rougeL_fmeasure',
 'rougeL_precision',
 'rougeL_recall',
 'rougeLsum_fmeasure',
 'rougeLsum_precision',
 'rougeLsum_recall']
for measure in rouges:
  df[measure] = df['rouge_scores'].apply(lambda x: x[measure].numpy().item())
df = df.drop(columns = ['rouge_scores'])

df.to_csv('/content/drive/MyDrive/MNLP Project/Data/results-final.csv', index = False)
```

**Hyperparameter Search**

```python
# lang_family_map = {'agg' : ['tamil', 'japanese', 'swahili'], 'fus': ['welsh','french', 'punjabi'],'iso' : ['burmese']}

# The path to the local git repo for Indic NLP Resources
INDIC_NLP_RESOURCES = r"indic_nlp_resources"

# Export the path to the Indic NLP Resources directory programmatically
common.set_resources_path(INDIC_NLP_RESOURCES)

# Initialize the Indic NLP library
loader.load()



def tokenize_japanese(text):
    mecab = MeCab.Tagger("-Owakati")
    return mecab.parse(text)

def tokenize_burmese(text):
    return " ".join(pds.tokenize(text))

def tokenize_tamil(text):
    return " ".join(indic_tokenize.trivial_tokenize(text))

def tokenize_welsh(text):
    tok = utokenize.Tokenizer(lang_code='cym')
    return tok.utokenize_string(text)

def tokenize_punjabi(text):
    return " ".join(indic_tokenize.trivial_tokenize(text))

def tokenize_french(text):
    tokens = nltk.word_tokenize(text, language='french')
    return " ".join(tokens)

def tokenize_swahili(text):
    # https://blog.neurotech.africa/swahili-text-classification-using-transformers/
    return nltk.word_tokenize(text)

tokenization_map = {'tamil' : tokenize_tamil, 'welsh' : tokenize_welsh, 'burmese' : tokenize_burmese, 'french' : tokenize_french, 'punjabi' : tokenize_punjabi, 'swahili' : tokenize_swahili, 'japanese' : tokenize_japanese}
```

```python
df_multilingual = pd.read_csv('/content/drive/MyDrive/MNLP Project/Data/results-final.csv')
languages = ['tamil', 'welsh', 'burmese', 'french', 'punjabi', 'swahili', 'japanese']
eval = {'monolingual' : {}, 'multilingual' : {}}
test = {'monolingual' : {}, 'multilingual' : {}}
results = {'tamil' : {'monolingual' : {}, 'multilingual' : {}}, 'welsh' : {'monolingual' : {}, 'multilingual' : {}}, 'burmese' : {'monolingual' : {}, 'multilingual' : {}}, 'french' : {'monolingual' : {}, 'multilingual' : {}}, 'punjabi' : {'monolingual' : {}, 'multilingual' : {}}, 'swahili' : {'monolingual' : {}, 'multilingual' : {}}, 'japanese' : {'monolingual' : {}, 'multilingual' : {}}}
comet_path = download_model("Unbabel/wmt22-comet-da")
comet = load_from_checkpoint(comet_path)

final_params = {}
final_eval_acc = {}


for language in languages:
  df_multi = pd.read_csv('/content/drive/MyDrive/MNLP Project/Data/' + language + '.csv')
  df_mono = pd.read_csv('/content/drive/MyDrive/MNLP Project/Data/' + language + '-monolingual.csv')

  eval['multilingual'][language], test['multilingual'][language] = train_test_split(df_multi, test_size=0.2, random_state = 69)
  eval['monolingual'][language], test['monolingual'][language] = train_test_split(df_mono, test_size=0.2, random_state = 69)
  gen_list_multi = eval['multilingual'][language].generated.tolist()
  gen_list_mono = eval['monolingual'][language].generated.tolist()
  gt_list = eval['multilingual'][language].ground_truth.tolist()

  comet_input_multi = []
  comet_input_mono = []
  for temp in zip(gt_list, gen_list_multi, gen_list_mono):
    comet_input_multi.append({'src': temp[0], 'mt' : temp[1], 'ref' : temp[0]})
    comet_input_mono.append({'src': temp[0], 'mt' : temp[2], 'ref' : temp[0]})
  results[language]['multilingual']['comet'] = comet.predict(comet_input_multi)['scores']
  results[language]['monolingual']['comet'] = comet.predict(comet_input_mono)['scores']

  rouge = ROUGEScore()
  results[language]['multilingual']['rouge'] = eval['multilingual'][language].apply(lambda x: rouge(x.generated, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
  results[language]['monolingual']['rouge'] = eval['monolingual'][language].apply(lambda x: rouge(x.generated, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()

  eval['multilingual'][language]['preprocessed'] = eval['multilingual'][language]['generated'].apply(lambda x: tokenization_map[language](x))
  eval['monolingual'][language]['preprocessed'] = eval['monolingual'][language]['generated'].apply(lambda x: tokenization_map[language](x))


  results[language]['multilingual']['rouge-pre'] = eval['multilingual'][language].apply(lambda x: rouge(x.preprocessed, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
  results[language]['monolingual']['rouge-pre'] = eval['monolingual'][language].apply(lambda x: rouge(x.preprocessed, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
for language in languages:
  best_params = {}
  best_acc = 0
  params = {}
  for _ in range(10):
    params['alpha'] = random.random()
    params['beta'] = random.random()
    params['gamma'] = random.random()

    multilingual_scores = [params['alpha'] * i + params['beta'] * j + params['gamma'] * k for i, j, k in zip(results[language]['multilingual']['rouge'], results[language]['multilingual']['rouge-pre'], results[language]['multilingual']['comet'])]
    monolingual_scores = [params['alpha'] * i + params['beta'] * j + params['gamma'] * k for i, j, k in zip(results[language]['monolingual']['rouge'], results[language]['monolingual']['rouge-pre'],results[language]['monolingual']['comet'])]
    acc = sum([i >= j for i, j in zip(multilingual_scores, monolingual_scores)])/len(monolingual_scores)
    print(f"Params - alpha: {params['alpha']}, beta: {params['beta']}, gamma: {params['gamma']}\nAccuracy: {acc}")
    if acc > best_acc:
      best_params = params
      best_acc = acc
  final_params[language] = best_params
  final_eval_acc[language] = best_acc
```

```python
test_results =  {'tamil' : {'monolingual' : {}, 'multilingual' : {}}, 'welsh' : {'monolingual' : {}, 'multilingual' : {}}, 'burmese' : {'monolingual' : {}, 'multilingual' : {}}, 'french' : {'monolingual' : {}, 'multilingual' : {}}, 'punjabi' : {'monolingual' : {}, 'multilingual' : {}}, 'swahili' : {'monolingual' : {}, 'multilingual' : {}}, 'japanese' : {'monolingual' : {}, 'multilingual' : {}}}
for language in languages:
  gen_list_multi = test['multilingual'][language].generated.tolist()
  gen_list_mono = test['monolingual'][language].generated.tolist()
  gt_list = test['multilingual'][language].ground_truth.tolist()

  comet_input_multi = []
  comet_input_mono = []
  for temp in zip(gt_list, gen_list_multi, gen_list_mono):
    comet_input_multi.append({'src': temp[0], 'mt' : temp[1], 'ref' : temp[0]})
    comet_input_mono.append({'src': temp[0], 'mt' : temp[2], 'ref' : temp[0]})
  test_results[language]['multilingual']['comet'] = comet.predict(comet_input_multi)['scores']
  test_results[language]['monolingual']['comet'] = comet.predict(comet_input_mono)['scores']

  rouge = ROUGEScore()
  test_results[language]['multilingual']['rouge'] = test['multilingual'][language].apply(lambda x: rouge(x.generated, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
  test_results[language]['monolingual']['rouge'] = test['monolingual'][language].apply(lambda x: rouge(x.generated, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()

  test['multilingual'][language]['preprocessed'] = test['multilingual'][language]['generated'].apply(lambda x: tokenization_map[language](x))
  test['monolingual'][language]['preprocessed'] = test['monolingual'][language]['generated'].apply(lambda x: tokenization_map[language](x))

  test_results[language]['multilingual']['rouge-pre'] = test['multilingual'][language].apply(lambda x: rouge(x.preprocessed, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
  test_results[language]['monolingual']['rouge-pre'] = test['monolingual'][language].apply(lambda x: rouge(x.preprocessed, x.ground_truth)['rougeLsum_fmeasure'].numpy().item(), axis = 1).tolist()
for language in languages:
  multilingual_scores = [final_params[language]['alpha'] * i + final_params[language]['beta'] * j + final_params[language]['gamma'] * k for i, j, k in zip(test_results[language]['multilingual']['rouge'], test_results[language]['multilingual']['rouge-pre'], test_results[language]['multilingual']['comet'])]
  monolingual_scores = [final_params[language]['alpha'] * i + final_params[language]['beta'] * j + final_params[language]['gamma'] * k for i, j, k in zip(test_results[language]['monolingual']['rouge'], test_results[language]['monolingual']['rouge-pre'], test_results[language]['monolingual']['comet'])]
  comet_acc = sum([i >= j for i, j in zip(test_results[language]['multilingual']['comet'], test_results[language]['monolingual']['comet'])])/len(monolingual_scores)
  rouge_acc = sum([i >= j for i, j in zip(test_results[language]['multilingual']['rouge'], test_results[language]['monolingual']['rouge'])])/len(monolingual_scores)
  acc = sum([i >= j for i, j in zip(multilingual_scores, monolingual_scores)])/len(monolingual_scores)
  print(f"Language: {language}\nNew Metric Accuracy: {acc}\nRouge Accuracy: {rouge_acc}\nComet Accuracy: {comet_acc}")
```
