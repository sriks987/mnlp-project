```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results-final-multilingual.csv")
df_2 = pd.read_csv("../results-final-monolingual.csv")
```

```python
df.head()
```

```python
df_grouped = df.groupby("language").mean()
df_grouped2 = df_2.groupby("language").mean()
```

```python
df_grouped.head(6)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming df_grouped and df_grouped2 are your dataframes
lang_col_names = df_grouped.columns[1:]

# Set the width of each bar
bar_width = 0.35

# Calculate the x-coordinates for the bars
categories = df_grouped.index
index = np.arange(len(categories))

for col_name in lang_col_names:
    # Plotting the side-by-side bar charts
    plt.bar(index, df_grouped[col_name], width=bar_width, label='Multilingual model')
    plt.bar(index + bar_width, df_grouped2[col_name], width=bar_width, label='Monolingual model')

    # Adding labels and title
    plt.xlabel('Categories')
    plt.ylabel(col_name)
    plt.title(f'Multilingual vs Monolingual performance for {col_name}')
    
    # Adding x-axis labels
    plt.xticks(index + bar_width / 2, categories)

    # Adding a legend
    plt.legend()

    # Display the plot
    plt.savefig(f'../results/full_res/{col_name}-bar.png', format = 'png')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming df_grouped and df_grouped2 are your dataframes
lang_col_names = df_grouped.columns[1:]

# Set the width of each bar
bar_width = 0.35

# Calculate the x-coordinates for the bars
categories = df_grouped.index
index = np.arange(len(categories))

filter_set = ['burmese', 'french', 'punjabi', 'swahili', 'tamil', 'welsh', 'japanese']

for filter_lang in filter_set:

    df_filtered = df[df['language'] == filter_lang]
    df2_filtered = df_2[df_2['language'] == filter_lang]

    for col_name in lang_col_names:
        # Plotting the side-by-side bar charts
        plt.boxplot([df_filtered[col_name], df2_filtered[col_name]], vert = True)
        plt.xticks([1,2], ['Multilingual model', 'Monolingual model'])
        plt.ylabel(f'{filter_lang} {col_name}')
        plt.xlabel('Model Type')

        plt.savefig(f'../results/{filter_lang}_res/{col_name}-{filter_lang}-box.png', format = 'png')
        plt.show()
```

```python
dic1 = {'french': 3547,
             'tamil': 591,
             'welsh': 65,
             'burmese': 207,
             'punjabi': 58,
             'swahili': 8}
dic2 = {'tamil': 2375,
 'welsh': 1218,
 'french': 1086,
 'punjabi': 1026,
 'swahili': 987,
 'burmese': 572}
dic2 = {'tamil': 2027,
             'welsh': 1216,
             'burmese': 570,
             'french': 1086,
             'punjabi': 1026,
             'swahili': 987,
             'japanese': 889}
dic3 = {'tamil': 591/17,
             'welsh': 65/17,
             'burmese': 207/17,
             'french': 3651/17,
             'punjabi': 58/17,
             'swahili': 8/17,
             'japanese': 468/17}

def sort_dict(myDict):
    myKeys = list(myDict.keys())
    myKeys.sort()
    sorted_dict = {i: myDict[i] for i in myKeys}
    return sorted_dict

dic1 = sort_dict(dic1)
dic2 = sort_dict(dic2)
dic3 = sort_dict(dic3)

def plot_dict(D):
    plt.bar(range(len(D)), list(D.values()), align='center')
    plt.xticks(range(len(D)), list(D.keys()))
    plt.title(f'Distribution of test samples')
    plt.ylabel(f'Number of test samples')

    plt.show()
    
plot_dict(dic1)
plot_dict(dic2)
plot_dict(dic3)
```
